```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def normalized_hit_rate(hit_rate):
    belady_raw = 0.3894
    lru_raw = 0.0440
    return ((hit_rate - lru_raw)/(belady_raw - lru_raw))
```

```python
models = [ "MLP (32)", "MLP (64)", "MLP (128)", "MLP (256)", "MLP (512)"]
hit_rates = np.array([ 0.2524, 0.3208, 0.0470, 0.0470, 0.0470])

normalized_hit_rates = normalized_hit_rate(hit_rates)

plt.figure(figsize=(9, 5))
bars = plt.bar(models, normalized_hit_rates, color = "green", width=0.5)

for bar, rate in zip(bars, normalized_hit_rates):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.02, 
             f"{rate:.2f}", ha='center', fontsize=11, fontweight='bold')

plt.ylabel("Normalized Cache Hit Rate")
plt.title("MLP: Normalized Cache Hit Rate vs. Perceptron Size")
plt.ylim(0, 1.1) 

plt.show()
```

```python
models = [ "RNN_w_Att (20)", "RNN_w_Att (40)", "RNN_w_Att (60)", "RNN_w_Att (80)", "RNN_w_Att (100)", "RNN_w_Att (120)", "RNN_w_Att (140)"]
hit_rates = np.array([ 0.1025, 0.3057, 0.3273, 0.3276, 0.3258, 0.3264, 0.3271])

normalized_hit_rates = normalized_hit_rate(hit_rates)

plt.figure(figsize=(8, 5))
bars = plt.bar(models, normalized_hit_rates, color="green",width=0.5 )

for bar, rate in zip(bars, normalized_hit_rates):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.02, 
             f"{rate:.2f}", ha='center', fontsize=10, fontweight='bold')

plt.ylabel("Normalized Cache Hit Rate")
plt.title("RNN With Attention: Normalized Cache Hit Rate vs. History Length ")
plt.ylim(0, 1.1) 
plt.xticks(rotation=50)
plt.show()
```

```python
models_wa = [ "RNN_w_Att (20)", "RNN_w_Att (40)", "RNN_w_Att (60)", "RNN_w_Att (80)", "RNN_w_Att (100)", "RNN_w_Att (120)", "RNN_w_Att (140)"]
hit_rates_wa = np.array([0.1025, 0.3057, 0.3273, 0.3276, 0.3258, 0.3264, 0.3271])

models_woa = [ "RNN_wo_Att (20)", "RNN_wo_Att (40)", "RNN_wo_Att (60)", "RNN_wo_Att (80)", "RNN_wo_Att (100)", "RNN_wo_Att (120)", "RNN_wo_Att (140)"]
hit_rates_woa = np.array([ 0.047, 0.047, 0.047, 0.047, 0.047, 0.047, 0.047])

normalized_with_attention  = normalized_hit_rate(hit_rates_wa)
normalized_without_attention = normalized_hit_rate(hit_rates_woa)

history_numeric = [20, 40, 60, 80, 100, 120, 140]  

# Plot
plt.figure(figsize=(8, 5))
plt.plot(history_numeric, normalized_with_attention, marker="o", linestyle="-", color="green", label="RNN With Attention")
plt.plot(history_numeric, normalized_without_attention, marker="s", linestyle="--", color="red", label="RNN Without Attention")

# Labels and title
plt.xlabel("History Length")
plt.ylabel("Normalized Cache Hit Rate")
plt.title("Normalized Cache Hit Rate vs. History Length (RNN w vs. w/o Attention)")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.7)

# Show plot
plt.show()
```
